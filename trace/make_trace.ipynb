```python
def is_vision_job(job):
    return job["model_type"] == "vision"
def is_nlp_job(job):
    return job["model_type"] == "transformer"
```

```python
def get_list_random_item(l):
    #np.random.seed(3)
    index = np.random.randint(0,len(l))
    #index = random.randint(0,len(l) - 1)# include the last element
    return l[index]
```

```python
def get_estimate_time(job):
    f = open("job_info.csv","r")
    d = csv.DictReader(f)

    ret_time = -1
    for e in d:
        if is_vision_job(job):
            if e["model"] == job["model_spec"]["model_name"] \
                and int(e["gpu_num"]) == job["worker_num"]:
                ret_time = float(e["estimated_iter_time/ms"])
        elif is_nlp_job(job):
            if e["model"] == "gpt-350M" \
                and int(e["gpu_num"].split("-")[0]) == job["parallel_spec"]["tp"] \
                and int(e["gpu_num"].split("-")[1]) == job["parallel_spec"]["pp"]:
                ret_time = float(e["estimated_iter_time/ms"])
    f.close()
    return ret_time
```

```python
def prefix_sums(lst:list):
    prefix_sum = []
    current_sum = 0
    for num in lst:
        current_sum += num
        prefix_sum.append(current_sum)
    #prefix_sum = [x - prefix_sum[0] for x in prefix_sum]
    return prefix_sum
```

```python
model_list = ["vgg16"] * 3 + ["resnet50"] + ["GPT-350M"] * 6 + ["mobilenet_v2"]
arrive_interval = 180

vision_bs = [32]
vision_dp = [1] * 3 + [2] * 2 + [4] * 2 + [8] * 4
vision_epochs = list(range(600,800))

nlp_bs = [(4,1)]
# [(tp, pp)]
nlp_parallel = [(2,1)] + [(1,2)] + \
               [(1,4)] + [(4,1)] * 3  + [(2,2)]  + \
               [(1,8)] + [(2,4)] + [(4,2)] + [(8,1)] * 7

nlp_iters = list(range(10000,20000))

job_num = 120
```

```python
def get_random_job(job_id, arrive_ts):
    new_job = dict()
    label = get_list_random_item(model_list)
    new_job["label"] = label
    new_job["id"] = job_id
    new_job["parallel_spec"] = dict()
    new_job["arrive_time"] = arrive_ts 
    new_job["model_spec"] = dict()
    new_job["jaca_score"] = 1000000#float("inf")

    if label == "GPT-350M":
        new_job["model_type"] = "transformer"
        new_job["parallel_spec"]["tp"], new_job["parallel_spec"]["pp"] = get_list_random_item(nlp_parallel)
        
        new_job["model_spec"]["iter_num"] = get_list_random_item(nlp_iters)
        if new_job["parallel_spec"]["tp"] == 8 or new_job["parallel_spec"]["tp"] == 4:
            new_job["model_spec"]["iter_num"] = int(new_job["model_spec"]["iter_num"] / 4)
        new_job["worker_num"] = new_job["parallel_spec"]["tp"] * new_job["parallel_spec"]["pp"]

        new_job["model_spec"]["iter_time"] = get_estimate_time(new_job)
        
    else:
        new_job["model_type"] = "vision"
        new_job["parallel_spec"]["dp"] = get_list_random_item(vision_dp)
        
        new_job["model_spec"]["epochs"] = get_list_random_item(vision_epochs)
        if label == "vgg16" and new_job["parallel_spec"]["dp"] > 2:
            new_job["model_spec"]["epochs"] = int(new_job["model_spec"]["epochs"] / 4)
            
        new_job["model_spec"]["model_name"] = label
        new_job["worker_num"] = new_job["parallel_spec"]["dp"]

        new_job["model_spec"]["epoch_time"] = get_estimate_time(new_job)

    return new_job
```

```python
def get_small_long_job(job_id, arrive_ts):
    new_job = dict()
    label = get_list_random_item(model_list)
    new_job["label"] = label
    new_job["id"] = job_id
    new_job["parallel_spec"] = dict()
    new_job["arrive_time"] = arrive_ts 
    new_job["model_spec"] = dict()
    new_job["jaca_score"] = 1000000#float("inf")

    if label == "GPT-350M":
        new_job["model_type"] = "transformer"
        new_job["parallel_spec"]["tp"], new_job["parallel_spec"]["pp"] = get_list_random_item(
             [(1,2)] * 2 + [(2,1)] * 4 + [(1,4)] + [(1,4)] + [(2,2)])
        
        new_job["model_spec"]["iter_num"] = get_list_random_item(nlp_iters) * 3
        new_job["worker_num"] = new_job["parallel_spec"]["tp"] * new_job["parallel_spec"]["pp"]

        new_job["model_spec"]["iter_time"] = get_estimate_time(new_job)
    else:
        new_job["model_type"] = "vision"
        new_job["parallel_spec"]["dp"] = get_list_random_item([1,1,2,2,4])
        
        new_job["model_spec"]["epochs"] = get_list_random_item(vision_epochs) * 3
        
        if new_job["parallel_spec"]["dp"] == 1:
            new_job["model_spec"]["epochs"] *= 3

        new_job["model_spec"]["model_name"] = label
        new_job["worker_num"] = new_job["parallel_spec"]["dp"]

        new_job["model_spec"]["epoch_time"] = get_estimate_time(new_job)

    return new_job
```

```python
def get_large_long_job(job_id, arrive_ts):
    new_job = dict()
    label = get_list_random_item(model_list)
    new_job["label"] = label
    new_job["id"] = job_id
    new_job["parallel_spec"] = dict()
    new_job["arrive_time"] = arrive_ts 
    new_job["model_spec"] = dict()
    new_job["jaca_score"] = 1000000#float("inf")

    if label == "GPT-350M":
        new_job["model_type"] = "transformer"
        new_job["parallel_spec"]["tp"], new_job["parallel_spec"]["pp"] = get_list_random_item(
            [(1,4)] + [(4,1)] + [(1,8)] + [(2,4)] * 2 + [(4,2)] * 2 + [(8,1)] * 2)
        
        new_job["model_spec"]["iter_num"] = get_list_random_item(nlp_iters) * 5
        new_job["worker_num"] = new_job["parallel_spec"]["tp"] * new_job["parallel_spec"]["pp"]

        new_job["model_spec"]["iter_time"] = get_estimate_time(new_job)
    else:
        new_job["model_type"] = "vision"
        new_job["parallel_spec"]["dp"] = get_list_random_item([4,8])
        
        new_job["model_spec"]["epochs"] = get_list_random_item(vision_epochs) * 5
            
        new_job["model_spec"]["model_name"] = label
        new_job["worker_num"] = new_job["parallel_spec"]["dp"]

        new_job["model_spec"]["epoch_time"] = get_estimate_time(new_job)

    return new_job
```
